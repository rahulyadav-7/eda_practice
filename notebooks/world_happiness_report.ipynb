```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
import seaborn as sns
%matplotlib inline
```

```python
# Some countries have longer names, lets fix that
data['Country'].replace("Palestinian Territories", "Palestine", inplace=True)
data['Country'].replace("Hong Kong S.A.R. of China", "Hong Kong", inplace=True)
data['Country'].replace("Taiwan Province of China", "Taiwan", inplace=True)
data['Country'] = data["Country"].str.replace("*","")
```

```python
# Rename column names
df=data.rename(columns = {'RANK':'Rank', 'Dystopia (1.83) + residual':'Dystopia', 'Explained by: GDP per capita':'GDP per capita', 
                     'Explained by: Social support':'Social support','Explained by: Healthy life expectancy':'Healthy life expectancy',
                        'Explained by: Freedom to make life choices':'Freedom to make life choices', 
                         'Explained by: Generosity':'Generosity', 'Explained by: Perceptions of corruption':'Perceptions of corruption'})
```

```python
# Use pairplot to have a look at the relationships among variables
sns.set_theme(style='ticks')
sns.pairplot(df)
```

```python
# Find numeric columns
num_cols = df.select_dtypes(include=['number']).columns.tolist()
num_cols
```

```python
# Let's have a heatmap to have a clear look on the relationships
corr = df[num_cols].corr()
mask = np.triu(np.ones_like(corr, dtype=bool))
f, ax = plt.subplots(figsize=(11,9))
cmap = sns.diverging_palette(230, 20, as_cmap=True)
sns.heatmap(corr, mask=mask, annot=True, cmap=cmap, vmax=.3, center=0, square=True, linewidths=.5, cbar_kws={'shrink':.5});
```

```python
# List top 10 happiest countries of the world
happiest_countries = df[['Country', 'Happiness score']].sort_values('Happiness score', ascending=False).head(10)
happiest_countries
```

```python
# Plotting countries on the map
import geopandas as gpd
# dataframe : happiest_countries

# Load the world map
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))

# Merge happiest_countries dataset with world geodataframe
world = world.merge(happiest_countries, how='left', left_on='name', right_on='Country')

# Plot the map
world.plot(column='Happiness score', cmap='Accent', legend=True, figsize=(15,10), missing_kwds={"color": "lightgrey", "label":"Missing values"})

plt.title('Top 10 Happiest Countries')
plt.show()
```
